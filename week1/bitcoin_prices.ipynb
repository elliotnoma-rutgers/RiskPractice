# download Bitcoin price history

```python
btc_df = yf.download('BTC-USD', 
                      start='2015-01-01', 
                      end='2025-01-22', 
                      progress=False, auto_adjust=True)
display(btc_df.head())
display(btc_df.tail())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,6))
plt.plot(btc_df['Close']['2022-01-20':], linewidth=3)
plt.title("Bitcoin price in USD", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.show()
```
