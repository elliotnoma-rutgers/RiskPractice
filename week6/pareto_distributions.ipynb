```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable

# Generate values for x-axis
x = np.linspace(-10, 10, 1000)

# Parameters for the Stable Paretian (Lévy) distribution
alpha = 1.5  # Stability parameter (0 < alpha <= 2)
beta = 0     # Skewness parameter (-1 <= beta <= 1)
scale = 1    # Scale parameter (> 0)
loc = 0      # Location parameter

params = [ (2.0, 0, 1, 0), (1.0, 0, 1, 0), (0.6, 0, 1, 0), (1.0, -0.99, 1, 0)]

# Plot the density curve
plt.figure(figsize=(8, 5))

for alpha, beta, scale, loc in params:
    # Compute probability density function (PDF)
    pdf_values = levy_stable.pdf(x, alpha, beta, loc=loc, scale=scale)
    plt.plot(x, pdf_values, label=f'Lévy Stable PDF (α={alpha}, β={beta})')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Stable Paretian (Lévy) Distribution')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

for alpha, beta, scale, loc in params:
    # Compute probability density function (PDF)
    pdf_values = levy_stable.cdf(x, alpha, beta, loc=loc, scale=scale)
    plt.plot(x, pdf_values, label=f'Lévy Stable PDF (α={alpha}, β={beta})')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Stable Paretian (Lévy) Distribution')
plt.legend()
plt.grid()
plt.show()
```
