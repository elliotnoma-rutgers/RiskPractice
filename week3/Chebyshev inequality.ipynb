```python
# https://stats.stackexchange.com/questions/235524/random-variables-for-which-markov-chebyshev-inequalities-are-tight
# http://www.ams.sunysb.edu/~jsbm/courses/311/cheby.pdf
# !pip install ipywidgets

import math
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def plot_Chebyshev(x, p, k, cheby_x, cheby_p, mean, sd, r, title_suffix):
    plt.bar(x, p, k/30, color='k')
    plt.plot(-cheby_x, cheby_p, '--')
    plt.title(f'{title_suffix} Chebyshev limit at {k:0.3}')
    plt.show()

    print(np.array([x, p]))
    print(f'{title_suffix} Chebyshev limit case mean = {mean:0.3}, sd = {sd:0.3}, r = {r:0.3}')

def plot_symmetric_Chebyshev_limit_distributions(k):
    k = float(k)
    x = [-k, 0, k]
    p = [1 / (2 * pow(k, 2)), 1 - 1 / pow(k, 2), 1 / (2 * pow(k, 2))]

    mean = sum([x0 * p0 for x0, p0 in zip(x, p)])
    sd = math.sqrt(sum([x0 * x0 * p0 for x0, p0 in zip(x, p)]))
    r = k / sd

    cheby_x = np.arange(.75, k + 1,.01)
    cheby_p = [0.5/(x * x) for x in cheby_x]
    plot_Chebyshev(x, p, k, cheby_x, cheby_p, mean, sd, r, "Symmetric")
```

Chebyshev Inequality
$$P(|X - \mu| \geq t) \leq \frac{\sigma^2}{t^2}$$

```python
k_slider = widgets.FloatSlider(
    min=1,
    max=5,
    description='r',
    step =0.25,
    value=1)

_ = interact(plot_symmetric_Chebyshev_limit_distributions, k=k_slider)
```

One-sided Chebyshev
$$P(X \leq \mu - a) \leq \frac{\sigma^2}{\sigma^2 + a^2}$$

```python
# https://stats.stackexchange.com/questions/235524/random-variables-for-which-markov-chebyshev-inequalities-are-tight
# Markov limit case

def plot_onesided_Chebyshev_limit_distributions(k):
    k = float(k)
    if k == 1:
        k = 1.01
    x = [-k, 0]
    p = [1 / pow(k, 2), 1 - 1 / pow(k, 2)]

    mean = np.array(sum([x0 * p0 for x0, p0 in zip(x, p)]))
    x -= mean
    mean = np.array(sum([x0 * p0 for x0, p0 in zip(x, p)]))
    sd = math.sqrt(sum([x0 * x0 * p0 for x0, p0 in zip(x, p)]))
    r = k / sd

    cheby_x = np.arange(.75, k + 1,.01)
    cheby_p = [sd * sd/(sd * sd + x * x) for x in cheby_x]
    plot_Chebyshev(x, p, k, cheby_x, cheby_p, mean, sd, r, "one sided")

_ = interact(plot_onesided_Chebyshev_limit_distributions, k=k_slider)
```
