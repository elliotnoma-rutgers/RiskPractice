```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
n = 250

phi = -0.5 # Autoregressive coefficient (controls how much the previous value influences cumulative sum
ar_white_noise = [np.random.normal(0,1)]
for t in range(1,n):
    new_value = phi * ar_white_noise[-1] + np.random.normal(0,1) # AR(1) process
    ar_white_noise.append(new_value)
ar_white_noise = np.array(ar_white_noise)
```

```python
series = ar_white_noise
plt.plot(series)
plt.title(f"AR1 white noise with phi = {phi}")
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
# Compute the Ljung-Box statistic using statsmodels
h = 15 # Maximum lag
ljung_box_results = acorr_ljungbox(series, lags=h, return_df=True)
print(f"Ljung-Box statistics and significance levels for AR1 white noise with phi = {phi}")
print(ljung_box_results)
```
