```python
import math
import numpy as np

def cholesky_equicorrelation(n, rho):
    # Initialize a zero matrix
    L = [[0.0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(i + 1):
            # Calculate the sum of squares of previous elements in the row
            sum_sq = sum(L[i][k] * L[j][k] for k in range(j))
            
            if i == j:
                # Diagonal elements: sqrt(A[i][i] - sum_sq)
                # For correlation matrix, A[i][i] is always 1
                L[i][j] = math.sqrt(1.0 - sum_sq)
            else:
                # Off-diagonal elements: (A[i][j] - sum_sq) / L[j][j]
                # For equicorrelation, A[i][j] is always rho
                L[i][j] = (rho - sum_sq) / L[j][j]
    return L

# Test with n=4, rho=0.5
n = 4
rho = 0.5
L = np.array(cholesky_equicorrelation(n, rho))

print(f"Cholesky Factor L (rho={rho}):")
print(L)
print()
print(np.matmul(L, L.T))
```

```python
from numpy.linalg import cholesky

rho = .4

corr = [[rho] * n for _ in range(n)]

for i in range(n):
    corr[i][i] = 1
print(np.array(corr))
print()

cholesky_decomp = cholesky(corr)
print(cholesky_decomp)
print()

print(np.matmul(cholesky_decomp, cholesky_decomp.T))
```
